Abstract classes are used as super classes to set the rules to its subclasses to implement the abstact mehtods in the super classes.

```python
from abc import abstractmethod
#program to illustrate abstract classes 
class Shape:
  @abstractmethod
  def area(self):
    pass
class Circle(Shape):
  def __init__(self,r):
    self.r=r
  def area(self):
    return 3.14 * self.r * self.r
class Rectangle(Shape):
  def __init__(self,l,b):
    self.l=l
    self.b=b
  def area(self):
    return self.l*self.b
c=Circle(1.5)
print(c.area())
r=Rectangle(1.2,3.5)
print(r.area())
```

Deriving a new class from more than one base class is known as multiple inheritance


```python
class C1:
  def m1(self):
    print('class1')
class C2:
  def m2(self):
    print('class2')
class C3(C1,C2):
  def m3(self):
    print('class3')
obj=C3()
obj.m1()
obj.m2()
obj.m3()  
```


decorators allows the programmer to modify  the behaviour  of a function .



```python
def decor(func):
  def inner(name):
     if  name =='Akash':
         print('Hai',name)
     else:
          func(name)
  return inner
@decor
def wish(name):
  print('Hello',name)
wish('Gaurav')
wish('Akash') 
wish('Ketan') 
```
